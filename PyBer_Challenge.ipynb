# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "resources/city_data.csv"
ride_data_to_load = "resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
# Create the Urban city DataFrame.
urban_cities_df= pyber_data_df[pyber_data_df["type"] == "Urban"]
urban_cities_df
```

```python
#Create the Suburban city DataFrames.
suburban_cities_df = pyber_data_df[pyber_data_df["type"] == "Suburban"]
suburban_cities_df.head()
```

```python
#Create the Rural city DataFrames.
rural_cities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]
rural_cities_df.head()
```

```python
#  1. Get the total rides for each city type
urban_ride_count=urban_cities_df["ride_id"].count()
suburban_ride_count=suburban_cities_df["ride_id"].count()
rural_ride_count=rural_cities_df["ride_id"].count()
rural_ride_count
```

```python
# 2. Get the total drivers for each city type
urban_drive_count=city_data_df.loc[(city_data_df.type=="Urban"),"driver_count"].sum()
suburban_drive_count=city_data_df.loc[(city_data_df.type=="Suburban"),"driver_count"].sum()
rural_drive_count=city_data_df.loc[(city_data_df.type=="Rural"),"driver_count"].sum()
rural_drive_count
```

```python
#  3. Get the total amount of fares for each city type
urban_fare_amount = urban_cities_df["fare"].sum()
suburban_fare_amount = suburban_cities_df["fare"].sum()
rural_fare_amount = rural_cities_df["fare"].sum()
rural_fare_amount
```

```python
#  4. Get the average fare per ride for each city type. 
urban_averagefare_ride=urban_fare_amount/urban_ride_count
suburban_averagefare_ride=suburban_fare_amount/suburban_ride_count
rural_averagefare_ride=rural_fare_amount/rural_ride_count

urban_averagefare_ride
```

```python
# 5. Get the average fare per driver for each city type. 
urban_averagefare_driver=urban_fare_amount/urban_drive_count
suburban_averagefare_driver=suburban_fare_amount/suburban_drive_count
rural_averagefare_driver=rural_fare_amount/rural_drive_count


rural_averagefare_driver
```

```python
#formatting dataframe

urban_ride_count=f"{urban_ride_count:,}"
suburban_ride_count=f"{suburban_ride_count:,}"
rural_ride_count=f"{rural_ride_count:,}"

urban_drive_count=f"{urban_drive_count:,}"
suburban_drive_count=f"{suburban_drive_count:,}"
rural_drive_count=f"{rural_drive_count:,}"

urban_fare_amount=f"${urban_fare_amount:5,.2f}"
suburban_fare_amount=f"${suburban_fare_amount:5,.2f}"
rural_fare_amount=f"${rural_fare_amount:5,.2f}"

urban_averagefare_ride=f"${'%.2f' % urban_averagefare_ride}"
suburban_averagefare_ride=f"${'%.2f' % suburban_averagefare_ride}"
rural_averagefare_ride=f"${'%.2f' % rural_averagefare_ride}"

urban_averagefare_driver=f"${'%.2f' % urban_averagefare_driver}"
suburban_averagefare_driver=f"${'%.2f' % suburban_averagefare_driver}"
rural_averagefare_driver=f"${'%.2f' % rural_averagefare_driver}"
```

```python
#  6. Create a PyBer summary DataFrame. 


data=np.array([["Urban",urban_ride_count,urban_drive_count,urban_fare_amount,urban_averagefare_ride,urban_averagefare_driver],
              ["Suburban",suburban_ride_count,suburban_drive_count,suburban_fare_amount,suburban_averagefare_ride,suburban_averagefare_driver],
              ["Rural",rural_ride_count, rural_drive_count,rural_fare_amount,rural_averagefare_ride,rural_averagefare_driver]])

pyber_summary_df = pd.DataFrame(data, columns = ["City type","Total Rides","Total Drivers","Total Fares", "Average fare per ride","Average fare per rider"])
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
pyber_summary_df=pyber_summary_df.sort_values(by=["Total Drivers"], ascending = False)
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
fares= pyber_data_df.groupby(["type","date"]).sum()["fare"]
fares_df= pd.DataFrame (fares)
fares_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
fares_df = fares_df.reset_index()
fares_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
fares_table= pd.pivot_table(fares_df,values=["fare"], columns=["type"], index=["date"])
fares_table.head(10)
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
new_table_df=fares_table.loc['2019-01-01':'2019-04-29']
new_table_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
new_table_df.index = pd.to_datetime(new_table_df.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
new_table_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week
table_resample_df=new_table_df.resample('W').sum()
table_resample_df.head(10)
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
table_resample_df.plot()
plt.style.use('fivethirtyeight')

plt.title("Total Fare by City Type", fontsize=16)
plt.ylabel("Fare ($USD)", fontsize=14)
fig=plt.figure()


plt.savefig("analysis/PyBer_fare_summary.png")

plt.show()
```
